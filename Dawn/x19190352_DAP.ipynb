Removing all rows that are empty or filled with NaN

```python
EduStatsNotEmpty = EduStats.dropna(how = "all")
```

```python
EduStatsNotEmpty
```

Removing bottom two rows also

```python
EduStatsNotEmpty = EduStatsNotEmpty.drop(index = [13135,13136])
EduStatsNotEmpty
```

Need to move the years into a single column and the series into multiple columns.  Will melt first.

```python
EduStatsMelt = EduStatsNotEmpty.melt(id_vars = ["Country_Name", "Country_Code", "Series", "Series_Code"], 
                                     var_name = "Year")
EduStatsMelt
```

Checking to make sure data has moved as expected

```python
EduStatsMelt.query('(`Country_Code` == "AFG") & (Year == "2010_YR2010")')
```

There are a lot of rows with ".." in them instead of values or NA need to change them as cannot pivot on non-numeric types

```python
EduStatsMelt = EduStatsMelt.replace({"..": NA})
EduStatsMelt["value"] = pd.to_numeric(EduStatsMelt["value"])
EduStatsMelt
```

```python
EduStatsMelt.query('(`Country_Code` == "AFG") & (Year == "2010_YR2010")')
```

```python
EduStatsPivot = EduStatsMelt.pivot_table(values = "value", index = ["Country_Name", "Country_Code", "Year"], 
                                   columns = ["Series"], aggfunc = np.sum)
EduStatsPivot
```

```python
EduStatsTable = EduStatsPivot.reset_index()             # Puts "Series" as a set of actual columns, not as the index
EduStatsTable
```

Checking for columns with only 0s in them

```python
(EduStatsTable != 0).any(axis=0)
```

```python
EduStatsTable.loc[:, (EduStatsTable != 0).any(axis=0)]
```

There are no columns with all zero entries, check for rows with all zero entries

```python
EduStatsTable = EduStatsTable[EduStatsTable.iloc[:, 4:].ne(0).any(axis=1)].reset_index(drop=True)
EduStatsTable
```

Normalizing the year column

```python
EduStatsTable['Year'] = EduStatsTable['Year'].str.split('_').str[0]
EduStatsTable
```

Will drop some columns so counting number of zero entries in each column to see which ones have the most data and are therefore the most useful

Trying some plots and visualisations

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (20,10))
ax = sns.boxplot(x="Year", y="Expenditure on education as % of total government expenditure (%)", data=EduStatsTable)
ax.set(xlabel = "Year", ylabel = "% Exp on Education")
```

```python
plt.figure(figsize= (20,10))
ax = sns.boxplot(x="Year", y="Gross enrolment ratio for tertiary education, female (%)", data=EduStatsTable)
ax.set(xlabel = "Year", ylabel = "Women in Third Level")
```

```python
plt.figure(figsize = (20,10))
ax = sns.boxplot(x="Year", y="Percentage of graduates from Science, Technology, Engineering and Mathematics programmes in tertiary education, both sexes (%)", data=EduStatsTable)
ax.set(xlabel = "Year", ylabel = "%STEM Graduates Tertiary")
```

```python
Countries2017Query = EduStatsTable.query('(Year == "2017")')
Countries2017EdPerGDP = Countries2017Query.loc[Countries2017Query['Government expenditure on education as % of GDP (%)']!= 0]
Countries2017EdPerGDP
```

```python
plt.figure (figsize = (30,15))
ax = sns.barplot(x = 'Country_Name', y = "Government expenditure on education as % of GDP (%)", data = Countries2017EdPerGDP)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.set(xlabel = "Country", ylabel = "Education % of GDP")
```

```python
#STEM Graduates in 2019 by country
STEMGrad2017 = Countries2017Query.loc[
    Countries2017Query['Percentage of graduates from tertiary education graduating from Natural Sciences, Mathematics and Statistics programmes, both sexes (%)']!= 0] 
plt.figure (figsize = (30,15))
ax = sns.barplot(x = 'Country_Name', y = "Percentage of graduates from tertiary education graduating from Natural Sciences, Mathematics and Statistics programmes, both sexes (%)", 
                 data = STEMGrad2017)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.set(xlabel = "Country", ylabel = "%STEM Graduates")
```

```python
EduStatsTenYears = EduStatsTable.query('(Year >= "2007") & (Year <= "2017")')
EduStatsTenYears
```

```python
EduStatsAverages = EduStatsTenYears.groupby('Country_Name').mean()
EduStatsAverages = EduStatsAverages.reset_index()
EduStatsAverages
```

```python
sns.pairplot(EduStatsAverages[['Government expenditure on education as % of GDP (%)',
                               'Percentage of graduates from tertiary education graduating from Information and Communication Technologies programmes, both sexes (%)',
                               'Percentage of graduates from tertiary education graduating from Natural Sciences, Mathematics and Statistics programmes, both sexes (%)',
                               'Percentage of graduates from tertiary education graduating from programmes in unspecified fields, both sexes (%)']], 
            diag_kind='hist')

plt.show()
```

```python
plt.figure (figsize = (30,15))
ax = sns.barplot(x = 'Country_Name', y = "Government expenditure on education as % of GDP (%)", data = EduStatsAverages)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.set(xlabel = "Country", ylabel = "%Expenditure of GDP")
```
